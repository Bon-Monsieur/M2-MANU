```python
%matplotlib widget

import matplotlib.pyplot as plt

from Lagrange_1D import Lagrange_1D
from Euler_1D import Euler_1D


def compar_eulerien_lagrangien(pb, flx_type, nc, cfl):
    
    plt.close('all')

    xs,den_sol,eni_sol,xi_eul,den_eul,eni_eul=Euler_1D(pb, flx_type, nc, cfl, 0)
    xs,den_sol,eni_sol,xi_lag,den_lag,eni_lag=Lagrange_1D(pb, flx_type, nc, cfl, 0)

    fig = plt.figure(1, figsize=(10, 8))
    plt.plot(xs, den_sol, '-r', linewidth=3)
    plt.plot(xi_lag, den_lag, '-dk')
    plt.plot(xi_eul, den_eul, '-*b')
    plt.xlabel('x')
    plt.ylabel('densité')
    plt.title("Densité à l'instant final")
    if (pb==1):
        cnter='upper right'
    else:
        cnter='upper center'
    plt.legend(['solution exacte', 'schéma Lagrangien', 'schéma Eulérien'], loc=cnter)

    fig2 = plt.figure(2, figsize=(10, 8))
    plt.plot(xs, eni_sol, '-r', linewidth=3)
    plt.plot(xi_lag, eni_lag, '-dk')
    plt.plot(xi_eul, eni_eul, '-*b')
    plt.xlabel('x')
    plt.ylabel('énergie interne')
    plt.title("Énergie interne à l'instant final")
    if (pb==1):
        cnter='upper left'
    else:
        cnter='upper right'
    plt.legend(['solution exacte', 'schéma Lagrangien', 'schéma Eulérien'], loc=cnter)

    plt.show()



## Voici un exemple de comment appeler la funtion
## compar_eulerien_lagrangien(1, 3, 100, 0.9)
```

```python
compar_eulerien_lagrangien(1, 3, 50, 0.9)
```
